**Machine learning _ Multiple Regression**



```python
df = pd.read_csv(r'APRICO2Decrease.csv') 
df = df.iloc[1:, :]
data = df.drop(['class name', 'people', 'start' , 'end', 'minimum', 'maximum'], axis='columns')
data.drop(data.columns[0], axis = 1, inplace = True ) # df.drop gives an error finding 'Date' so I delet by index
df_MR = data.dropna()
df_MR
```

```python
df_MR_target = df_MR.copy(deep=True)
df_MR_input = df_MR.copy(deep=True)
df_MR_target.drop(df_MR_target.columns[1:7], axis = 1, inplace = True)
df_MR_input.drop(df_MR_input.columns[0], axis = 1, inplace = True)
df_MR_input
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_MR_input,df_MR_target,test_size=0.2)
```

```python
m_reg_mod = linear_model.LinearRegression()
m_reg_mod.fit(X_train,y_train)
```

```python
print(m_reg_mod.coef_)
```

```python
pred = m_reg_mod.predict(X_test)
```

```python
predicted = m_reg_mod.predict([[50, 3.5, 4, 3, 2.5, 2]])
print(f' imaginary room {predicted}')
duplicate = m_reg_mod.predict([[50, 3.5, 4, 3, 2.5, 2]])
print(f' imaginary room {duplicate}')
predicted = m_reg_mod.predict([[50, 3.5, 4, 3, 2.5, 1]])
print(f' one door less {predicted}')
predicted = m_reg_mod.predict([[70, 3.5, 4, 2, 2.5, 2]])
print(f' bigger surface area {predicted}')
predicted = m_reg_mod.predict([[30, 3, 4, 3, 2.5, 2]])
print(f' smaller surface {predicted}')
```

    **Computing regression accuracy**
    
    Mean absolute error: This is the average of absolute errors of all the data points in the given dataset.
    
    Mean squared error: This is the average of the squares of the errors of all the data points in the given dataset. 
    It is one of the most popular metrics out there!
    
    Median absolute error: This is the median of all the errors in the given dataset. The main advantage of this 
    metric is that it's robust to outliers. A single bad point in the test dataset wouldn't skew the entire error  metric, as opposed to a mean error metric.
    
    Explained variance score: This score measures how well our model can account for the variation in our dataset. 
    A score of 1.0 indicates that our model is perfect.
    
    R2 score: This is pronounced as R-squared, and this score refers to the coefficient of determination.
    This tells us how well the unknown samples will be predicted by our model. The best possible score is 1.0,
    but the score can be negative as well.
    
    Functions ending with  _score return a value to maximize; the higher the better
    Functions ending with _error or _loss return a value to minimize; the lower the better

```python
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, pred), 2))
print("RMSE =", round(np.sqrt(mean_squared_error(y_test, pred)), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test, pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, pred), 2))
```
